```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")

data = pd.read_csv("chartData.csv")
data["total"] = data[data.columns[3:]].sum(axis=1)
data["young"] = data[data.columns[3:6]].sum(axis=1)
data["old"] = data[data.columns[17:]].sum(axis=1)
data["young%"] = data.young / data.total
data["old%"] = data.old / data.total

fig, ax = plt.subplots(figsize=(15, 10))
data[(data.year == 2058) & (data.sex == "p")].plot.barh(x="district", y="young%", width=.9, ax=ax)
```
